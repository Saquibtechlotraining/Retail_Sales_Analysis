```python
# Import all necessary python libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os                       # By using os module - We can fetch all the files name from any directory
```

# To load the data from  Sales.csv file, which we get by concat all the month-wise data One-by-one, and after we save this data as named: df


# Dealing with Missing Values

```python
df = pd.read_csv("SalesData/Sales.csv")
df
```

# Check Null values from all the columns of data set df:-


# finding info

```python
df.info()
```

# isnull()

```python
df.isnull()
```

```python
# Numbers of missing values (Null values) in all the columns of the data set df:-

df.isnull().sum()
```

# Delete Empty rows

```python
df = df.dropna(how = "all")
df
```

```python
# Checking that there will be no empty rows in the data set named: df

df.isnull().sum()
```

```python
# Checking that there will be no empty rows in the data set named: df  by doing the Above code Operation:-

df.info()
```

# Convert the datatype of  Quantity Ordered column from object (String) to integer.

```python
df["Quantity Ordered"].astype(int)
```

# Filter Out Wrong Data

```python
df.loc[df["Quantity Ordered"] == "Quantity Ordered"]    # We dont want this data because it is usefulless for Analysis
```

```python
# We want this usefull data by using loc Filtering:-

df = df.loc[df["Quantity Ordered"] != "Quantity Ordered"] 
df
```

# Resave

```python
df.to_csv("SalesData/Sales.csv")
```
