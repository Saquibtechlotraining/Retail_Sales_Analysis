# Task - Find Combination and then Find Combination count.

```python
# To Find, what Product purchase most with USB-C Charging Cable
# To Find, what Product purchase most with Wired Headphones
# To Find, what Product purchase most with Google Phone
# To Find, what Product purchase most with Vareebadd Phone
# To Find, what Product purchase most with Bose SoundSport Headphones
# etc in group column of gdf, we use combination Module

gdf["group"]   # This is series
```

```python
# Convert every values of group column in a list on the behalf of "," to Find the combination and Find combination Count.

gdf["group"] = gdf["group"].str.split(",")
gdf
```

```python
# To Find Combination and then Find Combination count.
# We use 2 modules :-

# 1) from itertools import combinations

# In the context of the, itertools module, the combinations function which provides a way to generate all possible combinations
# of a given length from an iterable.

# 2) from colllections import Counter

# In the context of the, collections module, the Counter is a built-in class that provides a convenient way to count the 
# occurrences of elements in a collection, such as a list, tuple, or string.
```

```python
# Here, we Find most_common combination Count:-

from itertools import combinations
from collections import Counter

count = Counter()      # Initiate Counter class in count variable

group = gdf["group"]

for sl in group:
    count.update(Counter(combinations(sl, 2)))  # use update function which is a part of count, to update  # update is non_return type function
    
d = dict(count.most_common(10))    # Here 10, denote 10 items
all_keys = list(d.keys())          
all_values = d.values()

di = {} # Empty dictionary
di["product"] = all_keys
di["count"] = all_values

di
```

```python
# Convert it into dictionary:-
# This is the Data set of  What products are most often sold together.

all_data = pd.DataFrame(di, columns=["product", "count"]).sort_values("count", ascending = False)
all_data
```
